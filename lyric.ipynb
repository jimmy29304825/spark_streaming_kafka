```python
# textblob score add into df
from textblob import TextBlob
# < 0: neg, >= 0: pos
for i in range(len(df)):
    df['textblob'][i] = TextBlob(str(df['lyric'].iloc[i])).sentiment[0]
df
```

```python
# afinn score add into df
from afinn import Afinn
# < 0: neg, >= 0: pos
afinn = Afinn()
df['Afinn'] = 0
for i in range(len(df)):
    l = str(df['lyric'].iloc[i])
    a = afinn.score(l)
    df['Afinn'][i] = a
df
```
