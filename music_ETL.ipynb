# 音樂分析
## 載入資料
## 資料過濾
    1. 選擇個人單首播放次數>=2的紀錄
    2. 篩選總聆聽次數>=20次的用戶
## 播放次數轉換成評分(1-5) - 百分位數
    1. 依據個人的播放紀錄定義0.2, 0.4, 0.6, 0.8百分位數之值並貼標
## 過濾出被播放次數最多的前200首歌
    將篩選結果與音樂詳細資料合併
   
## 使用評分跑推薦模型
    列出推薦順序丟給樹莓派

```python
import pandas as pd
import numpy as np
```

```python
# 載入資料
triplets_file = 'https://static.turi.com/datasets/millionsong/10000.txt'
song_df_1 = pd.read_table(triplets_file,header=None)
song_df_1.columns = ['user_id', 'song_id', 'listen_count']

songs_metadata_file = 'https://static.turi.com/datasets/millionsong/song_data.csv'
song_df_2 = pd.read_csv(songs_metadata_file)

# 將資料備份於本地端
song_df_1.to_csv('triplets_file.csv', encoding='utf-8', index=False)
song_df_2.to_csv('songs_metadata_file.csv', encoding='utf-8', index=False)
```

```python
# 資料過濾-1
## 篩選被個人播放次數大於等於2次的資料
song_df_1 = song_df_1[song_df_1['listen_count'] >= 2]
song_df_1
# song_df_1[song_df_1['listen_count'] >= 2]
```

```python
## 篩選總聆聽次數>=20次的用戶
tmp = song_df_1.groupby('user_id').sum()
tmp = tmp[tmp['listen_count'] >= 20]
tmp
```

```python
# 資料合併
res = pd.merge(song_df_1,tmp, on='user_id')
res.columns = ['user_id', 'song_id', 'listen_count', 'total_listen']
res
```

```python
# 建立用戶清單
user_id = res.drop_duplicates('user_id')['user_id']
user_id
```

```python
# 欄位格式轉換
pd.to_numeric(res['listen_count'])
pd.to_numeric(res['total_listen'])

# 新增欄位
res['level'] = 1.0
```

```python
# 依據個人的播放紀錄定義0.2, 0.4, 0.6, 0.8百分位數之值並貼標
res_new = pd.DataFrame(columns=('user_id', 'song_id', 'listen_count', 'total_listen'))
for x, i in enumerate(user_id):
    
    test = res[res['user_id'] == i]
#     print(len(test))
    five = test['listen_count'].quantile(0.8)
    four =test['listen_count'].quantile(0.6)
    three = test['listen_count'].quantile(0.4)
    two = test['listen_count'].quantile(0.2)
#     print(five, four, three, two, len(res[res['user_id'] == i]))
    # res[res['user_id'] == i]['listen_count'].iloc[0]
    for j in range(len(test)):

        if test['listen_count'].iloc[j] >= five:
            test['level'].iloc[j] = 5

        elif test['listen_count'].iloc[j] >= four:
            test['level'].iloc[j] = 4

        elif test['listen_count'].iloc[j] >= three:
            test['level'].iloc[j] = 3

        elif test['listen_count'].iloc[j] >= two:
            test['level'].iloc[j] = 2
            
    res_new = res_new.append(test, ignore_index=True)
    if x % 100 == 0:
        print(x, 'done')
res_new[['song_id', 'user_id', 'listen_count', 'level']]
```

```python
res_new = res_new[['song_id', 'user_id', 'listen_count', 'level']]
res_new
```

```python
# 資料另存
res_new.to_csv('music_rating.csv', encoding='utf-8', index=False)
```

```python
## 篩選總聆聽次數前200名的歌曲
tmp_song = res_new[['song_id', 'listen_count']].groupby('song_id').sum()
tmp_song = tmp_song.sort_values(['listen_count'], ascending = [0]).iloc[0:200]
tmp_song
```

```python
tmp_list = res_new.drop_duplicates('song_id')
tmp_song = pd.merge(tmp_list,tmp_song, on='song_id').sort_values(['listen_count_y'], ascending = [0])[['song_id','listen_count_y']]
```
